```python
crime_scat = df.plot('crim', 'lstat', kind='scatter', color='DarkBlue')
low_crime_rate = 5
df_low_crime = df.loc[(df['crim'] <= low_crime_rate)]
df_higher_crime = df.loc[(df['crim'] > low_crime_rate)]
df_low_crime.plot('crim', 'lstat', kind='scatter', color='DarkGreen', ax=crime_scat)
```

```python
crime_hist = df_higher_crime.plot('crim', 'lstat', kind='hist', color='DarkBlue')
low_crime_rate = 5
df_low_crime = df.loc[(df['crim'] <= low_crime_rate)]
df_low_crime.plot('crim', 'lstat', kind='hist', color='DarkGreen', ax=crime_hist, alpha=0.7)
```

Crime Per Capita Vs. % of Lower Status Members of Population

Crime is stereotypically accociated with poverty and "lower status" individuals, so I wanted to see if this held true for the Boston data set.  It does not.  The correlation for crime vs Lower Status is only 0.53 for the entire data set, only 0.24 for the middle 50 of Crime, and only 0.29 for the middle 50 of lower status.  The causation is something else besides the percentage of lower status people.

I split the data between lower crime and higher crime setting the division at 5 crimes per capita.  The lower crime data didn't add much of an explanation to the cause of crime, but when looking at the higher crime areas there were some interesting pieces of data.

In the higher crime areas, there was a static pupil/teacher ratio of 20 vs a range of 12.6 to 21.2 for the population having a mean of 18.45.  

In those same areas there was a static full value property tax per 10,000 of 666, where in the lower crime areas the tax values had a mean of 342 and the middle 50 percent falling between 269 and 399.  Why are the taxes so high in the higher crime area?  Those areas have a statis proportion of non-retail business acres of 18.1 while the lower crime areas have a mean of 9.5, about half.  As would be expected with the higher amounts of industry the higher crime areas generally have higher levels of polution.

The higher crime areas all have a 24 index of accessability where as the lower crime areas have a mean of 5.9.

Further investigation of what about the industrial zones may be a driver of crime is recommended.
